```python
def show_unique(column):
    base_unique_values = df[column].unique()
    cleaned_unique_values = df_cleaned[column].unique()
    print(f"Unique values in '{column}':")
    print(f"\nFOR Base DATASET:\n {base_unique_values}")
    print(f"{30 * '-'}")
    print(f"\nFOR CLEANED DATASET:\n {cleaned_unique_values}\n")
```

```python
attributes = ['room_type', 'city', 'host_identity_verified','host_response_rate' ]
for attr in attributes:
    print(f"{60 * '#'}\n")
    show_unique(attr)
```

```python
# Categorize review_scores_rating into three classes
def categorize_rating(rating):
    if rating >= 0 and rating < 3:
        return 'Low Rating'
    if rating >= 3 and rating < 4.51:
        return 'Medium Rating'
    elif rating >= 4.51 and rating < 4.91:
        return 'High Rating'
    elif rating >= 4.91 and rating <= 5:
        return 'Very High Rating'
    else:
        return None

df_cleaned['rating_category'] = df_cleaned['review_scores_rating'].apply(categorize_rating)
```

```python
# Show class distribution
print("Class Distribution:")
print("=" * 50)
print(df_cleaned['rating_category'].value_counts().sort_index())
print("\nPercentage Distribution:")
print("=" * 50)
print(df_cleaned['rating_category'].value_counts(normalize=True).sort_index() * 100)
```

```python
# OPTION 2: Tertile-based (33rd and 67th percentiles)
df_option2 = df_cleaned.copy()

q33 = df_option2['review_scores_rating'].quantile(0.33)
q67 = df_option2['review_scores_rating'].quantile(0.67)

def categorize_option2(rating):
    """Use tertiles for balanced classes"""
    if rating < q33:
        return 'Low'
    elif rating < q67:
        return 'Medium'
    else:
        return 'High'

df_option2['rating_category'] = df_option2['review_scores_rating'].apply(categorize_option2)

print("OPTION 2: Tertile-based (statistically balanced)")
print("=" * 60)
print(f"Thresholds: Low < {q33:.2f} | Medium: {q33:.2f}-{q67:.2f} | High >= {q67:.2f}")
print(f"Total rows: {len(df_option2)}\n")
print("Class Distribution:")
print(df_option2['rating_category'].value_counts().sort_index())
print("\nPercentage Distribution:")
print((df_option2['rating_category'].value_counts(normalize=True).sort_index() * 100).round(2))
```
